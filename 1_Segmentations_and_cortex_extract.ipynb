# Segmentations

```python
#defining directories/file names to search

out_dir = '/data/neuralabc/carfra/QuantMetComp/source/masks_created/'
in_dir = '/data/neuralabc/source/MPI_CBS/MPM_DTI/source/'

all_dirs = glob(in_dir+'*')
sub_ids = [os.path.basename(x) for x in all_dirs]
```

```python
def segmentations(in_dir, out_dir, sub_id):
    #segments a subject using mgdm
     
    out_dir = join(out_dir,sub_id,'MGDM')
    in_dir = join(in_dir,sub_id)
    
    if os.path.exists(join(out_dir,sub_id+'_mgdm-lbls.nii.gz')):
        a= 1
    else:
    
        FA = join(in_dir,sub_id+'_FA_reg.nii.gz')
        FA_ = nb.load(FA)

        R1 = join(in_dir,sub_id+'_R1.nii.gz')
        R11 = nb.load(R1)
        d1 = R11.get_data()
        d2 = np.zeros_like(d1)
        d2 = d1*1000     #R1 values are in the wrong range to work well with template, correct here
        R1_ = nb.Nifti1Image(d2,affine=R11.affine,header=R11.header)

        MT = join(in_dir,sub_id+'_MTsat.nii.gz')
        MT_ = nb.load(MT)

        PD = join(in_dir,sub_id+'_PDsat.nii.gz')
        PD_ = nb.load(PD)

        nighres.brain.mgdm_segmentation(contrast_image1 = FA_, contrast_type1 = 'DWIFA3T',
                                   contrast_image2 = R1_, contrast_type2 = 'mpmR13T',
                                   contrast_image3 = MT_, contrast_type3 = 'mpmMT3T',
                                   contrast_image4 = PD_, contrast_type4 = 'mpmPD3T',
                                   n_steps = 10,
                                   max_iterations = 1500,
                                   adjust_intensity_priors = False,
                                   normalize_qmaps = False,
                                   atlas_file = '/opt/quarantine/nighres/20190509/install/lib/python3.7/site-packages/nighres-1.1.0b1-py3.7.egg/nighres/atlases/brain-segmentation-prior3.0/brain-atlas-quant-3.0.8.txt',
                                   save_data = True, overwrite = True,
                                   output_dir = out_dir,
                                   file_name = sub_id)
```

```python
#Calling the function above with multiprocessing

pool = multiprocessing.ProcessingPool(nodes = 10)

seg = partial(segmentations, in_dir, out_dir)
res = pool.map(seg,sub_ids)

pool.close()
pool.clear()
```

# Cortex extraction

```python
#defining directories/file names to search

source_dir = '/data/neuralabc/source/MPI_CBS/MPM_DTI/source/'
seg_dir = '/data/neuralabc/source/MPI_CBS/MPM_DTI/processing/segmentations_MTSat_SPM/'
out_dir = '/data/neuralabc/carfra/QuantMetComp/source/masks_created/'

all_dirs = glob(source_dir+'*')
sub_ids = [os.path.basename(x) for x in all_dirs]
```

```python
def extract(out_dir, seg_dir, sub_id):   
    '''Uses nighres tools (cruise cortex extraction) to extract cortex 
    and create a levelset representation of the cortex. 
    Most of the code here is committed simply to finding the best masks to input to cruise cortex extraction (tested through trial and error).
    It is different than suggested in the documentation example for cortical depth estimation for this reason
    '''
    
    
    #csf mask (sulcal and ventricular)
    MD = nb.load(join('/data/neuralabc/source/MPI_CBS/MPM_DTI/source/',sub_id,sub_id+'_MD_reg.nii.gz'))
    csf_d = MD.get_data()
    csf_m = np.zeros_like(csf_d)
    csf_m[csf_d>0.0012] = 1
    csf_m[csf_d<0] = 1
    
    
    #WM mask (WM, subcortical GM, and ventricular CSF)
    MGDM_lbls = nb.load(join(out_dir,sub_id,'MGDM',sub_id+'_mgdm-lbls.nii.gz'))
    MGDM_lbls_data = MGDM_lbls.get_data()
    WM_mask = np.invert((MGDM_lbls_data==26)|(MGDM_lbls_data==27)|(MGDM_lbls_data==10)|(MGDM_lbls_data==20)|
                        (MGDM_lbls_data==50)|(MGDM_lbls_data==1)|(MGDM_lbls_data==25)|(MGDM_lbls_data==8)|
                        (MGDM_lbls_data==43)|(MGDM_lbls_data==46)|(MGDM_lbls_data==14))#subcortical tissues are 1
    WM_mask = WM_mask[:,:,:,0].astype(float)
    MGDM_data = nb.load(join(out_dir,sub_id,'MGDM',sub_id+'_mgdm-seg.nii.gz')).get_data()
    MGDM_WM_mask_rm = (MGDM_data==50)|(MGDM_data==1)|(MGDM_data==10)|(MGDM_data==25)|(MGDM_data==43)|(MGDM_data==46)|(MGDM_data==14)
    WM_fill = np.invert((MGDM_data==26)|(MGDM_data==27)|(MGDM_data==10)|(MGDM_data==20)|
                        (MGDM_data==50)|(MGDM_data==1)|(MGDM_data==25)|(MGDM_data==8)|
                        (MGDM_data==43)|(MGDM_data==46)|(MGDM_data==14)) #subcortical tissues are 1
    brain_to_erode = np.invert((MGDM_data==1)|(MGDM_data==43)|(MGDM_data==46)|(MGDM_data==25)|
                               (MGDM_data==20)|(MGDM_data==50)|(MGDM_data==14)|(MGDM_data==8))
    img = nb.Nifti1Image(brain_to_erode,affine=MGDM_lbls.affine,header=MGDM_lbls.header)
    img.to_filename(join(out_dir,sub_id,"MGDM",sub_id+'_to_erode.nii.gz'))   #keep! it is for the subprocess call below
    subprocess.call(["fslmaths",
                    join(out_dir,sub_id,"MGDM",sub_id+'_to_erode.nii.gz'),
                    "-ero",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz')])
    subprocess.call(["fslmaths",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz'),
                    "-ero",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz')])
    subprocess.call(["fslmaths",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz'),
                    "-ero",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz')])
    MGDM_ero = nb.load(join(out_dir,sub_id,'MGDM',sub_id+'_mgdm_eroded.nii.gz')).get_data()
    WM_mask_hard = (WM_mask*MGDM_ero).astype(bool)
    WM_mask_hard[WM_fill] = 1  #fills some ventricular CSF
    WM_hard = nb.Nifti1Image(WM_mask_hard,affine=MGDM_lbls.affine,header=MGDM_lbls.header)
    try:
        os.mkdir(join(out_dir,sub_id,"cortex_extraction"))
    except:
        print("Error creating: " + join(out_dir,sub_id,"cortex_extraction")+ "\n already existed")
    WM_hard.to_filename(join(out_dir,sub_id,"cortex_extraction",'WM_hard.nii.gz'))

    
    #sulcal CSF mask
    SPM_CSF = nb.load(join(seg_dir,sub_id,sub_id+'_CSF.nii.gz'))
    sulcal_CSF = SPM_CSF.get_data()
    sulcal_CSF[csf_m==1] = 1   ################
    sulcal_CSF[WM_mask_hard] = 0
    sulci_CSF = nb.Nifti1Image(sulcal_CSF,affine=SPM_CSF.affine,header=SPM_CSF.header)
#     sulci_CSF.to_filename(join(out_dir,sub_id,"cortex_extraction",'CSF.nii.gz'))

    
    #cortical GM mask
    SPM_GM = nb.load(join(seg_dir,sub_id,sub_id+'_GM.nii.gz'))
    MGDM_cort_mask = (MGDM_lbls_data==26)|(MGDM_lbls_data==27)#|(MGDM_data==10)|(MGDM_data==20)
    MGDM_cort_mask = MGDM_cort_mask[:,:,:,0]
#     cortical_GM = np.zeros_like(SPM_GM.get_data())
#     cortical_GM[MGDM_cort_mask] = SPM_GM.get_data()[MGDM_cort_mask]
    cortical_GM = SPM_GM.get_data()
    cortical_GM[WM_mask_hard] = 0
    cortical_GM[sulcal_CSF==1] = 0 #####
    GM = nb.Nifti1Image(cortical_GM,affine=SPM_GM.affine,header=SPM_GM.header)
#     GM.to_filename(join(out_dir,sub_id,"cortex_extraction",'GM.nii.gz'))



    nighres.cortex.cruise_cortex_extraction(init_image = WM_hard, 
                                            wm_image = WM_hard, 
                                            gm_image = GM, 
                                            csf_image = sulci_CSF, 
                                            vd_image=None, data_weight=0.4, 
                                            regularization_weight=0.1, max_iterations=500, 
                                            normalize_probabilities=False, correct_wm_pv=True, 
                                            wm_dropoff_dist=1.0, 
                                            topology='no', 
                                            topology_lut_dir=None, 
                                            save_data=True, 
                                            overwrite=True, 
                                            output_dir=join(out_dir,sub_id,"cortex_extraction"), 
                                            file_name=sub_id)
    
    #using the GM-WM boundary and GM-csf boundary to extract levelset of the middle of the cortex
    lss = [join(out_dir,sub_id,"cortex_extraction",sub_id+"_cruise-gwb.nii.gz"),
          join(out_dir,sub_id,"cortex_extraction",sub_id+"_cruise-cgb.nii.gz")]
    nighres.shape.levelset_fusion(levelset_images=lss, 
                                  correct_topology=True, 
                                  topology_lut_dir=None, 
                                  save_data=True, 
                                  overwrite=True, 
                                  output_dir=join(out_dir,sub_id,"cortex_extraction"), 
                                  file_name='cgb_gwb')
    #final levelset is the smallest of fused and 1.5 off gwb (to stay away from CSF partial voluming)
    fused_data = nb.load(join(out_dir,sub_id,"cortex_extraction","cgb_gwb_lsf-avg.nii.gz")).get_data()
    ls = nb.load(join(out_dir,sub_id,"cortex_extraction",sub_id+"_cruise-gwb.nii.gz"))
    ls_data = ls.get_data()-1.5
    ls_data[ls_data<fused_data] = fused_data[ls_data<fused_data]   
    new_ls = nb.Nifti1Image(ls_data,affine=ls.affine,header=ls.header)
    new_ls.to_filename(join(out_dir,sub_id,'cortex_ls.nii.gz'))

    #converting levelset of cortical sheath into a 
    probs = nighres.surface.levelset_to_probability(levelset_image = new_ls, 
                                            distance_mm=0.5, 
                                            save_data=False)
    GM_pr = probs['result'].get_data()
    GM_pr[GM_pr == 1] = 0
    GM_pr[GM_pr > 0] = 1
    GM_pr[MGDM_WM_mask_rm] = 0
    img = nb.Nifti1Image(GM_pr,affine=probs['result'].affine,header=probs['result'].header)
    img.to_filename(join(out_dir,sub_id,'cortex.nii.gz'))
```

```python
def extract_hemi(out_dir, seg_dir, sub_id):
    '''Uses nighres tools (cruise cortex extraction) to extract cortex (each hemisphere) 
    and create a levelset representation of the cortex. 
    Most of the code here is committed simply to finding the best masks to input to cruise cortex extraction (tested through trial and error).
    It is different than suggested in the documentation example for cortical depth estimation for this reason
    '''
    
    MD = nb.load(join('/data/neuralabc/source/MPI_CBS/MPM_DTI/source/',sub_id,sub_id+'_MD_reg.nii.gz'))
    csf_d = MD.get_data()
    csf_m = np.zeros_like(csf_d)
    csf_m[csf_d>0.0012] = 1
    csf_m[csf_d<0] = 1
    
    MGDM_lbls = nb.load(join(out_dir,sub_id,'MGDM',sub_id+'_mgdm-lbls.nii.gz'))
    MGDM_lbls_data = MGDM_lbls.get_data()
    WM_mask = np.invert((MGDM_lbls_data==26)|(MGDM_lbls_data==27)|(MGDM_lbls_data==10)|(MGDM_lbls_data==20)|
                        (MGDM_lbls_data==50)|(MGDM_lbls_data==1)|(MGDM_lbls_data==25)|(MGDM_lbls_data==8)|
                        (MGDM_lbls_data==43)|(MGDM_lbls_data==46)|(MGDM_lbls_data==14))
    WM_mask = WM_mask[:,:,:,0].astype(float)
    MGDM_data = nb.load(join(out_dir,sub_id,'MGDM',sub_id+'_mgdm-seg.nii.gz')).get_data()
    MGDM_WM_mask_rm = (MGDM_data==50)|(MGDM_data==1)|(MGDM_data==10)|(MGDM_data==25)|(MGDM_data==43)|(MGDM_data==46)|(MGDM_data==14)
    WM_fill = np.invert((MGDM_data==26)|(MGDM_data==27)|(MGDM_data==10)|(MGDM_data==20)|
                        (MGDM_data==50)|(MGDM_data==1)|(MGDM_data==25)|(MGDM_data==8)|
                        (MGDM_data==43)|(MGDM_data==46)|(MGDM_data==14))
    brain_to_erode = np.invert((MGDM_data==1)|(MGDM_data==43)|(MGDM_data==46)|(MGDM_data==25)|
                               (MGDM_data==20)|(MGDM_data==50)|(MGDM_data==14)|(MGDM_data==8))
    img = nb.Nifti1Image(brain_to_erode,affine=MGDM_lbls.affine,header=MGDM_lbls.header)
    img.to_filename(join(out_dir,sub_id,"MGDM",sub_id+'_to_erode.nii.gz'))   #keep! it is for the subprocess call below
    subprocess.call(["fslmaths",
                    join(out_dir,sub_id,"MGDM",sub_id+'_to_erode.nii.gz'),
                    "-ero",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz')])
    subprocess.call(["fslmaths",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz'),
                    "-ero",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz')])
    subprocess.call(["fslmaths",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz'),
                    "-ero",
                    join(out_dir,sub_id,"MGDM",sub_id+'_mgdm_eroded.nii.gz')])
    MGDM_ero = nb.load(join(out_dir,sub_id,'MGDM',sub_id+'_mgdm_eroded.nii.gz')).get_data()
    WM_mask_hard = (WM_mask*MGDM_ero).astype(bool)
    mask_l = ((MGDM_data==48)|(MGDM_data==37)|(MGDM_data==33)|(MGDM_data==41)|(MGDM_data==27)|(MGDM_data==10)|(MGDM_data==50)|(MGDM_data==18)|
             (MGDM_data==39)|(MGDM_data==31)|(MGDM_data==35)|(MGDM_data==12)|(MGDM_data==27)|(MGDM_data==8)|(MGDM_data==13))
    mask_r = ((MGDM_data==47)|(MGDM_data==36)|(MGDM_data==32)|(MGDM_data==40)|(MGDM_data==26)|(MGDM_data==10)|(MGDM_data==50)|(MGDM_data==17)|
             (MGDM_data==38)|(MGDM_data==30)|(MGDM_data==34)|(MGDM_data==11)|(MGDM_data==26)|(MGDM_data==8)|(MGDM_data==13))
    WM_mask_hard[WM_fill] = 1
    WM_mask_hard[MGDM_WM_mask_rm] = 0
    WM_mask_hard_l = np.copy(WM_mask_hard)
    WM_mask_hard_l[mask_r] = 0
    WM_mask_hard_r = np.copy(WM_mask_hard)
    WM_mask_hard_r[mask_l] = 0
    try:
        os.mkdir(join(out_dir,sub_id,"cortex_extraction"))
    except:
        print("Error creating: " + join(out_dir,sub_id,"cortex_extraction")+ "\n already existed")
    WM_hard_l = nb.Nifti1Image(WM_mask_hard_l,affine=MGDM_lbls.affine,header=MGDM_lbls.header)
    WM_hard_l.to_filename(join(out_dir,sub_id,"cortex_extraction",'WM_hard_l.nii.gz'))
    WM_hard_r = nb.Nifti1Image(WM_mask_hard_r,affine=MGDM_lbls.affine,header=MGDM_lbls.header)
    WM_hard_r.to_filename(join(out_dir,sub_id,"cortex_extraction",'WM_hard_r.nii.gz'))

    SPM_CSF = nb.load(join(seg_dir,sub_id,sub_id+'_CSF.nii.gz'))
    sulcal_CSF_l = SPM_CSF.get_data()
    sulcal_CSF_l[csf_m==1] = 1
    sulcal_CSF_l[WM_mask_hard] = 0
    sulcal_CSF_r = np.copy(sulcal_CSF_l)
    sulcal_CSF_l[mask_r] = 1
    sulcal_CSF_r[mask_l] = 1
    sulci_CSF_l = nb.Nifti1Image(sulcal_CSF_l,affine=SPM_CSF.affine,header=SPM_CSF.header)
    sulci_CSF_r = nb.Nifti1Image(sulcal_CSF_r,affine=SPM_CSF.affine,header=SPM_CSF.header)
#     sulci_CSF.to_filename(join(out_dir,sub_id,"cortex_extraction",'CSF.nii.gz'))

    SPM_GM = nb.load(join(seg_dir,sub_id,sub_id+'_GM.nii.gz'))
    MGDM_cort_mask = (MGDM_lbls_data==26)|(MGDM_lbls_data==27)#|(MGDM_data==10)|(MGDM_data==20)
    MGDM_cort_mask = MGDM_cort_mask[:,:,:,0]
#     cortical_GM = np.zeros_like(SPM_GM.get_data())
#     cortical_GM[MGDM_cort_mask] = SPM_GM.get_data()[MGDM_cort_mask]
    cortical_GM_l = SPM_GM.get_data()
    cortical_GM_l[WM_mask_hard] = 0
    cortical_GM_l[csf_m==1] = 0
    cortical_GM_r = np.copy(cortical_GM_l)
    cortical_GM_l[mask_r] = 0
    cortical_GM_r[mask_l] = 0
    GM_l = nb.Nifti1Image(cortical_GM_l,affine=SPM_GM.affine,header=SPM_GM.header)
    GM_r = nb.Nifti1Image(cortical_GM_r,affine=SPM_GM.affine,header=SPM_GM.header)
#     GM.to_filename(join(out_dir,sub_id,"cortex_extraction",'GM.nii.gz'))

    cort = nighres.cortex.cruise_cortex_extraction(init_image = WM_hard_l, 
                                                   wm_image = WM_hard_l, 
                                                   gm_image = GM_l, 
                                                   csf_image = sulci_CSF_l, 
                                                   vd_image=None, data_weight=0.4, 
                                                   regularization_weight=0.1, max_iterations=500, 
                                                   normalize_probabilities=False, correct_wm_pv=True, 
                                                   wm_dropoff_dist=1.0, 
                                                   topology='no', 
                                                   topology_lut_dir=None, 
                                                   save_data=True, 
                                                   overwrite=True, 
                                                   output_dir=join(out_dir,sub_id,"cortex_extraction_l"), 
                                                   file_name=sub_id)
    lss = [join(out_dir,sub_id,"cortex_extraction_l",sub_id+"_cruise-gwb.nii.gz"),
          join(out_dir,sub_id,"cortex_extraction_l",sub_id+"_cruise-cgb.nii.gz")]
    nighres.shape.levelset_fusion(levelset_images=lss, 
                                  correct_topology=True, 
                                  topology_lut_dir=None, 
                                  save_data=True, 
                                  overwrite=True, 
                                  output_dir=join(out_dir,sub_id,"cortex_extraction_l"), 
                                  file_name='cgb_gwb')
    
    cort = nighres.cortex.cruise_cortex_extraction(init_image = WM_hard_r, 
                                                   wm_image = WM_hard_r, 
                                                   gm_image = GM_r, 
                                                   csf_image = sulci_CSF_r, 
                                                   vd_image=None, data_weight=0.4, 
                                                   regularization_weight=0.1, max_iterations=500, 
                                                   normalize_probabilities=False, correct_wm_pv=True, 
                                                   wm_dropoff_dist=1.0, 
                                                   topology='no', 
                                                   topology_lut_dir=None, 
                                                   save_data=True, 
                                                   overwrite=True, 
                                                   output_dir=join(out_dir,sub_id,"cortex_extraction_r"), 
                                                   file_name=sub_id)
    lss = [join(out_dir,sub_id,"cortex_extraction_r",sub_id+"_cruise-gwb.nii.gz"),
          join(out_dir,sub_id,"cortex_extraction_r",sub_id+"_cruise-cgb.nii.gz")]
    nighres.shape.levelset_fusion(levelset_images=lss, 
                                  correct_topology=True, 
                                  topology_lut_dir=None, 
                                  save_data=True, 
                                  overwrite=True, 
                                  output_dir=join(out_dir,sub_id,"cortex_extraction_r"), 
                                  file_name='cgb_gwb')
    
    fused_data_l = nb.load(join(out_dir,sub_id,"cortex_extraction_l","cgb_gwb_lsf-avg.nii.gz")).get_data()
    ls_l = nb.load(join(out_dir,sub_id,"cortex_extraction_l",sub_id+"_cruise-gwb.nii.gz"))
    ls_data_l = ls_l.get_data()-1.5
    ls_data_l[ls_data_l<fused_data_l] = fused_data_l[ls_data_l<fused_data_l]
    new_ls_l = nb.Nifti1Image(ls_data_l,affine=ls_l.affine,header=ls_l.header)
    new_ls_l.to_filename(join(out_dir,sub_id,'cortex_ls_l.nii.gz'))
    fused_data_r = nb.load(join(out_dir,sub_id,"cortex_extraction_r","cgb_gwb_lsf-avg.nii.gz")).get_data()
    ls_r = nb.load(join(out_dir,sub_id,"cortex_extraction_r",sub_id+"_cruise-gwb.nii.gz"))
    ls_data_r = ls_r.get_data()-1.5
    ls_data_r[ls_data_r<fused_data_r] = fused_data_r[ls_data_r<fused_data_r]
    new_ls_r = nb.Nifti1Image(ls_data_r,affine=ls_r.affine,header=ls_r.header)
    new_ls_r.to_filename(join(out_dir,sub_id,'cortex_ls_r.nii.gz'))

    probs_l = nighres.surface.levelset_to_probability(levelset_image = new_ls_l, 
                                            distance_mm=0.5, 
                                            save_data=False)
    probs_r = nighres.surface.levelset_to_probability(levelset_image = new_ls_r, 
                                            distance_mm=0.5, 
                                            save_data=False)

    full = nb.load(join(out_dir,sub_id,'cortex.nii.gz')).get_data()
    
    GM_pr = probs_l['result'].get_data()
    GM_pr[GM_pr == 1] = 0
    GM_pr[GM_pr > 0] = 1
    GM_pr[MGDM_WM_mask_rm] = 0
    GM_pr[full==0] = 0
    img = nb.Nifti1Image(GM_pr,affine=probs_l['result'].affine,header=probs_l['result'].header)
    img.to_filename(join(out_dir,sub_id,'cortex_l.nii.gz'))
    GM_pr = probs_r['result'].get_data()
    GM_pr[GM_pr == 1] = 0
    GM_pr[GM_pr > 0] = 1
    GM_pr[MGDM_WM_mask_rm] = 0
    GM_pr[full==0] = 0
    img = nb.Nifti1Image(GM_pr,affine=probs_r['result'].affine,header=probs_r['result'].header)
    img.to_filename(join(out_dir,sub_id,'cortex_r.nii.gz'))
```
